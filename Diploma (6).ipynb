### График корреляции переменных
Отдельно для непрерывных переменных - корреляция Пирсона, для ординальных - Спирмена (можно и для нерперывных - The Spearman's test can be used to analyse ordinal level, as well as continuous level data, because it uses ranks instead of assumptions of normality.), для бинарных - корреляция через В Крамера. Пример для Крамера тут: https://www.kaggle.com/code/chrisbss1/cramer-s-v-correlation-matrix/notebook
https://medium.com/@ktoprakucar/how-to-calculate-the-correlation-between-categorical-and-continuous-values-dcb7abf79406

```python
from scipy.stats import chi2_contingency
import numpy as np

#Создаем функцию для расчета Ви Крамера
def cramers_V(var1,var2) :
  crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
  stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
  obs = np.sum(crosstab) # Number of observations
  mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
  return np.sqrt(stat/(obs*mini))


data_encoded = df.drop(columns=['jobstatus','identificator18','identificator19','identificator20','identificator21','identificator22','year'])

rows= []
for var1 in data_encoded:
  col = []
  for var2 in data_encoded :
    cramers =cramers_V(data_encoded[var1], data_encoded[var2]) # Cramer's V test
    col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
  rows.append(col)
  
cramers_results = np.array(rows)
#Создаем матрицу корреляции
df3 = pd.DataFrame(cramers_results, columns = data_encoded.columns, index =data_encoded.columns)
```

```python
# Отсюда получаем, что для модели отдельно надо включать (лучше только наличие включать, кол-во портит модель):
# либо кол-во детей, либо их наличие
# либо кол-во подчиненных, либо их наличие
# либо кол-во часов работы из дома, либо наличие работы из дома
# либо кол-во дней отпуска, либо его наличие
# либо имт, либо его соответствие норме
df_bin = df3.drop(df3.loc[['age','jobsatisfaction','conditions_satisf', 'wage_satisf', 'opport_satisf','subordinates_num', 'workhours','workfromhome_hours', 'wage','vacation_duration','wealth_status', 'authority_status',
       'respect_status','children_number', 'work_experience_years', 'health', 'whenstartsmoke','imt']].index)
fig = plt.figure(figsize=(10, 5))
ax = sns.heatmap(df_bin,cmap="crest")

plt.show()
```

```python
#Отсюда получаем, чтов модель одновременно нельзя включать возраст и опыт работы
df_cont = df[['age','jobsatisfaction','conditions_satisf', 'wage_satisf', 'opport_satisf','subordinates_num', 'workhours','workfromhome_hours', 'wage','vacation_duration','wealth_status', 'authority_status',
       'respect_status','children_number', 'work_experience_years', 'health', 'whenstartsmoke','imt']]
sns.heatmap(df_cont.corr(method = 'spearman'),cmap="crest")
```
